```python
import pandas as pd
import matplotlib.pyplot as plt


base_path = '/Users/gabeschwartz/desktop/final_project/data_raw/'  

races = pd.read_csv(base_path + "races.csv")
results = pd.read_csv(base_path + "results.csv")
drivers = pd.read_csv(base_path + "drivers.csv")
circuits = pd.read_csv(base_path + "circuits.csv")


race_results = pd.merge(races, results, on='raceId')
race_driver_results = pd.merge(race_results, drivers, on='driverId')
final_data = pd.merge(race_driver_results, circuits, left_on='circuitId', right_on='circuitId')

# Create a home/away flag
```
